```python
combined_df['daily_percentage'] = 0
```

```python
combined_df.loc[combined_df['Ticker'].values == "TSLA"]
```

```python
tickers = ["SPY", "BND", "TSLA"]
for ticker in tickers:
    combined_df['daily_percentage'].loc[combined_df['Ticker'].values == ticker] = combined_df['Close'].loc[combined_df['Ticker'].values == ticker].pct_change()
```

```python
combined_df['daily_percentage'] = combined_df['daily_percentage'].fillna(0)
```

```python
combined_df
```

```python
combined_df.loc[combined_df['Ticker'].values == "SPY"]
```

```python
plt.figure(figsize=(18, 6))  # Set the figure size for better readability
sns.lineplot(combined_df,x=combined_df.index, y='daily_percentage', hue="Ticker")    # Plot the DataFrame; each column is a line

    # Add title and labels
plt.title('Daily Returns Over Time', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Daily Returns', fontsize=12)
# Customize the legend and grid
plt.legend(title='Ticker')
plt.grid(True)
    # Adjusts plot to ensure everything fits without overlapping

plt.show()
```

```python
combined_df['rolling_mean'] = 0
```

```python
combined_df.loc[combined_df['rolling_mean'].isnull()]
```

```python
for ticker in tickers:
    combined_df['rolling_mean'].loc[combined_df['Ticker'].values == ticker] = combined_df['daily_percentage'].loc[combined_df['Ticker'].values == ticker].rolling(window=30).mean()
```

```python
combined_df['rolling_std'] = 0
```

```python
for ticker in tickers:
    combined_df['rolling_std'].loc[combined_df['Ticker'].values == ticker] = combined_df['daily_percentage'].loc[combined_df['Ticker'].values == ticker].rolling(window=30).std()
```

```python
combined_df['rolling_mean'] = combined_df['rolling_mean'].fillna(0)
combined_df['rolling_std'] = combined_df['rolling_std'].fillna(0)
```

## Rolling Average mean

```python
plt.figure(figsize=(18, 6))  # Set the figure size for better readability
sns.lineplot(combined_df,x=combined_df.index, y='rolling_mean', hue="Ticker")    # Plot the DataFrame; each column is a line

    # Add title and labels
plt.title('Daily Returns Over Time', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Daily Returns', fontsize=12)
# Customize the legend and grid
plt.legend(title='Ticker')
plt.grid(True)
    # Adjusts plot to ensure everything fits without overlapping

plt.show()
```

## Rolling Average Standard deviation

```python
plt.figure(figsize=(18, 6))  # Set the figure size for better readability
sns.lineplot(combined_df,x=combined_df.index, y='rolling_std', hue="Ticker")    # Plot the DataFrame; each column is a line

    # Add title and labels
plt.title('Daily Returns Over Time', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Daily Returns', fontsize=12)
# Customize the legend and grid
plt.legend(title='Ticker')
plt.grid(True)
    # Adjusts plot to ensure everything fits without overlapping

plt.show()
```
