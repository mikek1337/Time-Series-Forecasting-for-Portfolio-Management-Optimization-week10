### Perform ADF test
to determine if the time series is stationary or not

performing the test on the closing price

```python
from statsmodels.tsa.stattools import adfuller
for ticker in tickers:
    result_prices = adfuller(combined_df['Close'].loc[combined_df['Ticker'].values == ticker])
    print(f"ADF testing result for {ticker} closing price ")
    print(f"ADF Stats {result_prices[0]}")
    print(f"p-value {result_prices[1]}")
```

preforming the test on daily returns

```python
for ticker in tickers:
    result_prices = adfuller(combined_df['daily_percentage'].loc[combined_df['Ticker'].values == ticker])
    print(f"ADF testing result for {ticker} daily returns ")
    print(f"ADF Stats {result_prices[0]}")
    print(f"p-value {result_prices[1]}")
```

```python
confidence_level = 0.05

daily_return = combined_df['daily_percentage'].loc[combined_df['Ticker'] == "TSLA"]
VaR = np.percentile(daily_return, confidence_level)
print("--- Key Risk Metrics for Tesla (TSLA) ---")
print(f"95% Value at Risk (VaR): {VaR:.4f}")
print("This means you are 95% confident that the daily loss will not exceed this percentage.")
print("\n" + "="*40 + "\n")
```

```python
risk_free_rate = 0.03
trading_days_per_year = 252

mean_daily_return = daily_return.mean()
std_daily_return = daily_return.std()

annual_return = mean_daily_return * trading_days_per_year
annual_std = std_daily_return * np.sqrt(trading_days_per_year)

# Calculate the Sharpe Ratio
sharpe_ratio = (annual_return - risk_free_rate) / annual_std

print("--- Sharpe Ratio ---")
print(f"Annualized Return: {annual_return:.4f}")
print(f"Annualized Volatility (Std Dev): {annual_std:.4f}")
print(f"Sharpe Ratio: {sharpe_ratio:.4f}")
print("A higher Sharpe Ratio indicates a better return for the level of risk taken.")
```
